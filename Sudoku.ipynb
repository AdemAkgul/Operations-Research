```python
import gurobipy as gb
```

```python
mdl = gb.Model("Sudoku")
```

```python
n = 16 #dimension of grid
t = int(n**0.5)#dimension of subgrid
```

```python
x = mdl.addVars(n,n,n, vtype = gb.GRB.BINARY, name = "Xijk")#if number k is assigned to row i and row j, it equals to 1, ow 0
```

```python
mdl.addConstrs((gb.quicksum(x[i,j,k]  for k in range(n)) == 1  
                for i in range(n) for j in range(n)),name = "s")#every cell must include one value
```

```python
mdl.addConstrs((gb.quicksum(x[i,j,k]  for j in range(n)) == 1  
                 for i in range(n) for k in range(n)),name = "s1")#every row must include a distinct number
```

```python
mdl.addConstrs((gb.quicksum(x[i,j,k]  for i in range(n)) == 1  
                 for j in range(n) for k in range(n)),name = "s2")#every column must include a distinct number
```

```python
mdl.addConstrs((gb.quicksum(x[i+m,j+l,k] for i in range(t) for j in range(t)) == 1  
                  for m in range(0,n,t) for l in range(0,n,t) for k in range(n)),name = "s1")#every t x t subgrid must include a distinct value
```

```python
mdl.params.Heuristics = 0.5
mdl.params.PreSolve = 2
mdl.optimize()
```

```python
mdl.printAttr('X')
```

```python
assignment = [(i,j,k) for i,j,k in x if x[i,j,k].x == 1]
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.zeros((n,n)))
for i,j,k in assignment:
    df.iloc[i,j] = k
```
