Every task is assigned to exactly one worker in this assignment problem

```python
import gurobipy as gb
import numpy as np
import pandas as pd
import time
```

```python
n = 1000#number of workers and tasks. They are equal in this case.
A = np.random.randint(2,10000,(n,n))          
C = [(i,j) for i in range(n) for j in range(n)]
```

```python
mdl = gb.Model("Assignment")
```

```python
a = time.time()
y = mdl.addVars(C, vtype = gb.GRB.CONTINUOUS, name = "Yij")#if task is assigned to worker j, it Yij equals to1, ow 0
b = time.time()
print(b-a,"elapsed time (second)")
```

```python
a = time.time()
mdl.setObjective(gb.quicksum(y[i,j]*A[i,j] for i,j in C),gb.GRB.MINIMIZE)
b = time.time()
print(b-a,"elapsed time (second)")
```

```python
a = time.time()
mdl.addConstrs(gb.quicksum(y[i,j] for j in range(n)) == 1 for i in range(n))
mdl.addConstrs(gb.quicksum(y[i,j] for i in range(n)) == 1 for j in range(n))
b = time.time()
print(b-a,"elapsed time (second)")
```

```python
mdl.Params.Method = 1#for speeding up the model
mdl.optimize()
```
